Попробуем модель glove-twitter-25 - Предварительно обученные GloVe векторы, основанные на 2B твитах, 27B токенах, 1,2 млн слов, без кавычек.

```python
glove_vectors = gensim.downloader.load('glove-twitter-25')
```

## 2. Тестирование модели <a class="anchor" id="second-bullet"></a>

Посмотрим на расстояние от слова Paris до разных слов из check_words

```python
check_words = ['spain','table','paris','idea','state','coffee','bird','moon', 'stone', 'eiffel']
```

```python
for i in check_words:
    similaruty = glove_vectors.similarity('france', i)
    print(f'расстояние между france и {i}: {similaruty}')
```
