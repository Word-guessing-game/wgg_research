## 1. Загрузка модели и получение эмбеддингов слов <a class="anchor" id="first-bullet"></a>

Загружаем модель

```python
cap_path = datapath("crime-and-punishment.bin")
wv = load_facebook_vectors(cap_path)
```

Какие слова есть в корпусе:

```python
len(wv.key_to_index)
```

```python
wv.key_to_index
```

## 2. Тестирование модели <a class="anchor" id="second-bullet"></a>

Посмотрим на расстояние между эммбедингом слова "france" и произвольными словами, чтобы оценить качество эмбеддингов

```python
check_words = ['spain','table','paris','idea','state','coffee','bird','moon', 'stone', 'eiffel']
```

```python
for i in check_words:
    similaruty = wv.similarity('france', i)
    print(f'расстояние между france и {i}: {similaruty}')
```

Вывод: Судя по всем, лучше подобрать другую предобученную модель, где будут только английские слова и больше слов
